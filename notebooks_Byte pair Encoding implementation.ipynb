```python
import numpy as np
from collections import Counter

# -------------------------------
# Corpus (No corpus.txt required)
# -------------------------------
words = [
    "low",
    "lower",
    "lowest",
    "new",
    "newer",
    "newest",
    "wide",
    "wider",
    "widest"
]

print("Corpus:")
print(words)

# -------------------------------
# Create Vocabulary
# -------------------------------
vocabulary = Counter()

for word in words:
    token = " ".join(list(word)) + " </w>"
    vocabulary[token] += 1

print("\n========== Initial Vocabulary ==========")

for word, freq in vocabulary.items():
    print(word, ":", freq)

frequencies = np.array(list(vocabulary.values()))

print("\nWord Frequencies:", frequencies)
print("Total Words:", np.sum(frequencies))
print("Maximum Frequency:", np.max(frequencies))


# -------------------------------
# Count Pairs
# -------------------------------
def count_pairs(vocab):
    pairs = Counter()

    for word, freq in vocab.items():
        symbols = word.split()

        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i + 1])
            pairs[pair] += freq

    return pairs


# -------------------------------
# Merge Pair
# -------------------------------
def merge_pair(pair, vocab):

    new_vocab = {}

    bigram = " ".join(pair)
    replacement = "".join(pair)

    for word in vocab:
        new_word = word.replace(bigram, replacement)
        new_vocab[new_word] = vocab[word]

    return new_vocab


# -------------------------------
# Learn Merge Rules
# -------------------------------
merge_rules = []

num_merges = 10

for i in range(num_merges):

    pairs = count_pairs(vocabulary)

    if len(pairs) == 0:
        break

    best_pair = max(pairs, key=pairs.get)

    merge_rules.append(best_pair)

    print("\n--------------------------")
    print("Merge", i + 1)
    print("Best Pair :", best_pair)
    print("Frequency :", pairs[best_pair])

    vocabulary = merge_pair(best_pair, vocabulary)


# -------------------------------
# Final Vocabulary
# -------------------------------
final_vocab = set()

for word in vocabulary:
    final_vocab.update(word.split())

print("\n========== Final Vocabulary ==========")

print(sorted(final_vocab))

print("Vocabulary Size :", len(final_vocab))


# -------------------------------
# Encode
# -------------------------------
def encode(word, merge_rules):

    tokens = list(word) + ["</w>"]

    for pair in merge_rules:

        i = 0

        while i < len(tokens) - 1:

            if tokens[i] == pair[0] and tokens[i + 1] == pair[1]:

                tokens = tokens[:i] + ["".join(pair)] + tokens[i + 2:]

            else:
                i += 1

    return tokens


# -------------------------------
# Decode
# -------------------------------
def decode(tokens):

    word = "".join(tokens)

    return word.replace("</w>", "")


# -------------------------------
# Testing
# -------------------------------
test_words = [
    "newest",
    "lowest",
    "lower"
]

print("\n========== Testing ==========")

for word in test_words:

    encoded = encode(word, merge_rules)

    decoded = decode(encoded)

    print("\nOriginal :", word)
    print("Encoded  :", encoded)
    print("Decoded  :", decoded)


# -------------------------------
# Merge Rules
# -------------------------------
print("\n========== Learned Merge Rules ==========")

for i, rule in enumerate(merge_rules, start=1):
    print(i, ":", rule)

print("\n========== Program Completed ==========")
```
